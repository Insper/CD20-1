## Esperança e Variância de uma variável aleatória $X$

<div id="discreta"></div>

## Discreta:

Seja $X$ uma variável aleatória **discreta** que assume valores $x$'s e 
$P(X=x)$ sendo função de probabilidade.

A esperança e a variância de $X$  são dadas por:

### Esperança:

$\qquad \mu = E(X) = \sum_{x} x P(X=x)$

### Variância: 

$\qquad Var(X) = \sum_{x}\left(x - \mu\right)^2 P(X=x)$


$\quad$
<div id="continua"></div>

## Contínua:

Seja $X$ uma variável aleatória **contínua** que assume valores $x$'s e 
$f(x)$ sendo função densidade de probabilidade.

A esperança e a variância de $X$ são dadas por:

### Esperança:

$\qquad \mu = E(X) = \int_{-\infty}^{\infty}x f(x) dx$

### Variância:

$\qquad Var(X) = \int_{-\infty}^{\infty}\left(x - \mu\right)^2 f(x) dx$


$\quad$

## Alternativamente, variância de $X$:

$Var(X) = E(X^2) - \left[E(X)\right]^2 = E(X^2) - \mu^2$

<div id='const'></div>

## Propriedades para $X+d$ e para $cX$

Seja $X$ uma variável aleatória qualquer e $c$ e $d$ constantes.

As esperanças e variâncias de $X+d$ e de $cX$ são dadas por:

<div id="espconst"></div>

### Esperança:

$\qquad E(X+d) = E(X) + d$

$\qquad E(cX) = cE(X)$

Combinando as expressões acima:

$\qquad E(cX + d) = cE(X) + d$


<br>
<div id="varconst"></div>

### Variância:

$\qquad Var(X+d) = Var(X)$

$\qquad Var(cX) = c^2Var(X)$

Combinando as expressões acima:

$\qquad Var(cX + d) = c^2Var(X)$

<div id='comb1'></div>

## Propriedades para $X+Y$ e para $X-Y$

Sejam $X$ e $Y$ variáveis aleatórias quaisquer.

As esperanças e variâncias de $X+Y$ e de $X-Y$ são dadas por:

<div id="esp1"></div>

### Esperança:

$\qquad E(X + Y) = E(X) + E(Y)$

$\qquad E(X - Y) = E(X) - E(Y)$

Estes resultados são sempre válidos, sejam $X$ e $Y$ dependentes ou independentes 


<br>
<div id="var1"></div>

### Variância:

$\qquad Var(X + Y) = Var(X) + Var(Y) + 2Cov(X,Y)$

$\qquad Var(X - Y) = Var(X) + Var(Y) - 2Cov(X,Y)$


<br>
<div id="indep1"></div>

### Variáveis independentes

Se $X$ e $Y$ são independentes $\Rightarrow$ $Cov(X,Y)=0$.

Entretanto, vale lembrar que 
se $Cov(X,Y)=0$ $\nRightarrow$ $X$ e $Y$ sejam independentes.

Se as variáveis $X$ e $Y$ forem independentes, então as variâncias simplificam para:

$\qquad Var(X+Y) = Var(X) + Var(Y)$

$\qquad Var(X-Y) = Var(X) + Var(Y)$



___

## Propriedades para $aX+bY$ e para $aX-bY$

Sejam $X$ e $Y$ variáveis aleatórias quaisquer e $a$ e $b$ sendo constantes.

As esperanças e variâncias de $aX+bY$ e de $aX-bY$ são dadas por:

<div id="esp2"></div>

### Esperança:

$\qquad E(aX + bY) = aE(X) + bE(Y)$

$\qquad E(aX - bY) = aE(X) - bE(Y)$

Estes resultados são sempre válidos, sejam $X$ e $Y$ dependentes ou independentes 


<br>
<div id="var2"></div>

### Variância:

$\qquad Var(aX + bY) = a^2Var(X) + b^2Var(Y) + 2ab Cov(X,Y)$

$\qquad Var(aX - bY) = a^2Var(X) + b^2Var(Y) - 2ab Cov(X,Y)$


<br>
<div id="indep2"></div>

### Variáveis independentes

Se $X$ e $Y$ são independentes $\Rightarrow$ $Cov(X,Y)=0$.

Entretanto, vale lembrar que 
se $Cov(X,Y)=0$ $\nRightarrow$ $X$ e $Y$ sejam independentes.

Se as variáveis $X$ e $Y$ forem independentes, então as variâncias anteriores simplificam para:

$\qquad Var(a^2X + b^2Y) = a^2Var(X) + b^2Var(Y)$

$\qquad Var(a^2X - b^2Y) = a^2Var(X) + b^2Var(Y)$

## Propriedades para $X_1+X_2+X_3$

Sejam $X_1$, $X_2$ e $X_3$ variáveis aleatórias quaisquer.

A esperança e a variância de $X_1 + X_2 + X_3$ são dadas por:

<div id="esp3"></div>

### Esperança:

$\qquad E(X_1 + X_2 + X_3) = E(X_1) + E(X_2) + E(X_3)$

Este resultado é sempre válido, sejam $X_i$'s dependentes ou independentes entre si.


<br>
<div id="var3"></div>

### Variância:

$\qquad Var(X_1+X_2+X_3) = Var(X_1)+Var(X_2)+Var(X_3)+2Cov(X_1,X_2)+2Cov(X_1,X_3)+2Cov(X_2,X_3)$


<br>
<div id="indep3"></div>

### Variáveis independentes

Se $X_i$'s são todos independentes entre si, então este resultado simplifica para:

$\qquad Var(X_1+X_2+X_3) = Var(X_1)+Var(X_2)+Var(X_3)$

<div id='soma'></div>


## Propriedade em Soma de variáveis independentes

Sejam $X_1, X_2,\ldots,X_n$  variáveis aleatórias quaisquer com $E(X_i) = \mu_i$ e $Var(X_i) = \sigma_i^2$, para $i=1,2,\ldots, n$.
Ainda, considere que $c_i$'s sejam constantes.

Definindo a soma de variáveis independentes:

$\qquad Y = c_1X_1 + c_2X_2 + \ldots + c_nX_n$

então a esperança e a variância de $Y$ são dadas por:

$\qquad E(Y) = c_1\mu_1 + c_2\mu_2 + \ldots + c_n\mu_n$

$\qquad Var(Y) = c_1^2\sigma_1^2 + c_2^2\sigma_2^2 + \ldots + c_n^2\sigma_n^2$


<br>

### Resultado importante:

Se $X_i$'s são normais independentes, então $Y$ **será uma variável aleatória normalmente distribuída**, com $E(Y)$ e $Var(Y)$ definidas conforme acima.

___
<div id='media'></div>

##  Propriedade em Média de variáveis independentes

Sejam $X_1, X_2,\ldots,X_n$  variáveis aleatórias quaisquer independentes com mesma média e mesma variância, ou seja, com $E(X_i) = \mu$ e $Var(X_i) = \sigma^2$, para $i=1,2,\cdots, n$.

Definindo a média dessas variáveis independentes por:

$\qquad \overline{X} = \frac{(X_1 + X_2 + \cdots + X_n)}{n}$

então a esperança e a variância de $\overline{X}$ são dadas por:

$\qquad E(\overline{X}) = \mu$

$\qquad Var(\overline{X}) = \frac{\sigma^2}{n}$


<br>

### Resultado importante:

Se $X_i$'s são independentes e identicamente distribuídas a normal, então $\overline{X}$ **será uma variável aleatória normalmente distribuída**, com $E(\overline{X})$ e $Var(\overline{X})$ definidas conforme acima.
